```python
# Problem 1: Social Media Relationship (9 points)

# Let us pretend that you are building a new app. Your app supports social media functionality, which means that users can have relationships with other users.

# There are two guidelines for describing relationships on this social media app:
# 1. Any user can follow any other user.
# 2. If two users follow each other, they are considered friends.


# Write a function called relationship_status that takes three positional arguments: (str) from_member, (str) to_member, and (dict) social_graph. The function should determine the relationship status of the first member to the second member based on the data stored in the social graph. The function should return one of these values depending on what is appropriate:

# "follower", if from_member follows to_member
# "followed by", if from_member is followed by to_member
# "friends", if from_member and to_member follow each other
# None if none of the above scenarios are applicable
# Specifications:

# from_member and to_member are user IDs (e.g. "@jobenilagan").
# social_graph is a dictionary that adheres to the specification demonstrated in the demo cell.

# NON-INTERACTIVE CODE CELL. YOU MAY RUN THIS CELL, BUT DO NOT EDIT IT.
# FOR DEMONSTRATION PURPOSES ONLY. DO NOT EDIT.
    
social_graph = {
    "@bongolpoc":{"first_name":"Joselito",
                  "last_name":"Olpoc",
                  "following":[
                  ]
    },
    "@joaquin":  {"first_name":"Joaquin",
                  "last_name":"Gonzales",
                  "following":[
                      "@chums","@jobenilagan"
                  ]
    },
    "@chums" : {"first_name":"Matthew",
                "last_name":"Uy",
                "following":[
                    "@bongolpoc","@miketan","@rudyang","@joeilagan"
                ]
    },
    "@jobenilagan":{"first_name":"Joben",
                   "last_name":"Ilagan",
                   "following":[
                    "@eeebeee","@joeilagan","@chums","@joaquin"
                   ]
    },
    "@joeilagan":{"first_name":"Joe",
                  "last_name":"Ilagan",
                  "following":[
                    "@eeebeee","@jobenilagan","@chums"
                  ]
    },
    "@eeebeee":  {"first_name":"Elizabeth",
                  "last_name":"Ilagan",
                  "following":[
                    "@jobenilagan","@joeilagan"
                  ]
    },
}
    
    
def relationship_status(from_member, to_member, social_graph):
    people_following_from_member = social_graph[str(from_member)]['following']
    if str(to_member) not in social_graph.keys(): # Considering the cases of '@miketan' and '@rudyang'
        people_following_to_member = social_graph[str(from_member)]['following']
        return 'follower'
    else: # If the user_IDs are keys within the dictionary
        people_following_to_member = social_graph[str(to_member)]['following']
        if (str(from_member) not in social_graph.keys()) or (str(to_member) not in social_graph.keys()):
            return 'Please enter a valid account name.'
        elif str(from_member) == str(to_member):
            return 'Please enter two different account names.'
        elif (str(to_member) in people_following_from_member) and (str(from_member) in people_following_to_member):
            return 'friends'
        elif str(to_member) in people_following_from_member:
            return 'follower'
        elif str(from_member) in people_following_to_member:
            return 'followed by'
        else:
            return None

    
# Sample Inputs
                            # from       # to                                # relationship_status
print(relationship_status('@bongolpoc','@joaquin', social_graph))            # None
print(relationship_status('@bongolpoc','@chums', social_graph))              # followed by
print(relationship_status('@bongolpoc','@jobenilagan', social_graph))        # None
print(relationship_status('@bongolpoc','@joeilagan', social_graph))          # None
print(relationship_status('@bongolpoc','@eeebeee', social_graph))            # None
print(relationship_status('@bongolpoc','@bongolpoc', social_graph))          # Two different account names
print(relationship_status('@joaquin','@bongolpoc', social_graph))            # None
print(relationship_status('@joaquin','@chums', social_graph))                # follower
print(relationship_status('@joaquin','@jobenilagan', social_graph))          # friends
print(relationship_status('@joaquin','@joeilagan', social_graph))            # None
print(relationship_status('@joaquin','@eeebeee', social_graph))              # None
print(relationship_status('@jobenilagan','@bongolpoc', social_graph))        # None
print(relationship_status('@jobenilagan','@joaquin', social_graph))          # friends
print(relationship_status('@jobenilagan','@chums', social_graph))            # follower
print(relationship_status('@jobenilagan','@joeilagan', social_graph))        # friends
print(relationship_status('@jobenilagan','@eeebeee', social_graph))          # friends
print(relationship_status('@joeilagan','@bongolpoc', social_graph))          # None
print(relationship_status('@joeilagan','@joaquin', social_graph))            # None
print(relationship_status('@joeilagan','@chums', social_graph))              # friends
print(relationship_status('@joeilagan','@jobenilagan', social_graph))        # friends
print(relationship_status('@joeilagan','@eeebeee', social_graph))            # friends
print(relationship_status('@eeebeee','@bongolpoc', social_graph))            # None
print(relationship_status('@eeebeee','@joaquin', social_graph))              # None
print(relationship_status('@eeebeee','@chums', social_graph))                # None
print(relationship_status('@eeebeee','@jobenilagan', social_graph))          # friends
print(relationship_status('@eeebeee','@joeilagan', social_graph))            # friends
print(relationship_status('@chums','@bongolpoc', social_graph))              # follower
print(relationship_status('@chums','@joaquin', social_graph))                # followed by
print(relationship_status('@chums','@jobenilagan', social_graph))            # followed by
print(relationship_status('@chums','@joeilagan', social_graph))              # friends
print(relationship_status('@chums','@eeebeee', social_graph))                # None
print(relationship_status('@chums','@miketan', social_graph))                # follower
print(relationship_status('@chums','@rudyang', social_graph))                # follower
```

```python
# NEW APPROACH

# Problem 2: Noughts and Crosses (10 points)

# Noughts and Crosses is a common paper-and-pencil game. 
# Players must attempt to successfully draw a straight line of their symbol across a grid. 
# The player that does this first is considered the winner.


# Write a function called tic_tac_toe that takes one positional argument (list) board. The function should return the winner of the board if there is one, and it should return None if there is no winner on the board.

# Specifications:

# 1. Each player is represented by their symbol. Example: if the player using 'X' won the board, then simply return the string 'X'.
# 2. The board may be 3x3, 4x4, 5x5, or 6x6.
# 3. The game will only ever be player X vs player O. No other symbols will be used.
# 4. There may be no winner, and there may be 1 winner, but there will never be a situation where there is more than 1 winner.

# NON-INTERACTIVE CODE CELL. YOU MAY RUN THIS CELL, BUT DO NOT EDIT IT.
# FOR DEMONSTRATION PURPOSES ONLY. DO NOT EDIT.


# Board 1: 'X'
board1 = [
['X','X','O'],
['O','X','O'],
['O','','X'],
]

# Board 2: 'X'
board2 = [
['X','X','O'],
['O','X','O'],
['','O','X'],
]

# Board 3: 'O'
board3 = [
['O','X','O'],
['','O','X'],
['X','X','O'],
]

# Board 4: 'X'
board4 = [
['X','X','X'],
['O','X','O'],
['O','','O'],
]

# Board 5: 'O'
board5 = [
['X','X','O'],
['O','X','O'],
['X','','O'],
]

# Board 6: 'None'
board6 = [
['X','X','O'],
['O','X','O'],
['X','',''],
]

# Board 7: 'O'
board7 = [
['','X','O','X'],
['X','O','O','X'],
['O','X','X','O'],
['O','O','O','O'],
]

# Board 8: 'O'
board8 = [
['','X','O','X','X'],
['X','O','O','X','X'],
['O','X','X','O','X'],
['O','O','O','O','O'],
['O','X','X','O','X'],
]

# Board 9: 'X'
board9 = [
['','X','O','X','X','O'],
['X','O','O','X','X','X'],
['O','X','X','O','X','O'],
['O','O','O','O','X','X'],
['O','X','X','O','X',''],
['O','X','X','O','X',''],
]


def tic_tac_toe(board):
    if all([board[len(board)-i-1][i] == 'O' for i in range(len(board))]): # Diagonal Upwards
        return 'O'
    elif all([board[len(board)-i-1][i] == 'X' for i in range(len(board))]): # Diagonal Upwards
        return 'X'
    elif all ([board[i][i] == 'O' for i in range(len(board))]): # Diagonal Downwards
        return 'O'
    elif all ([board[i][i] == 'X' for i in range(len(board))]): # Diagonal Downwards
        return 'X'
    elif any([line == ['O'] * len(board) for line in board]): # Horizontal Combination
        return 'O'
    elif any([line == tuple(['O'] * len(board)) for line in zip(*board)]): # Vertical Combination
        return 'O'
    elif any([line == ['X'] * len(board) for line in board]): # Horizontal Combination
        return 'X'
    elif any([line == tuple(['X'] * len(board)) for line in zip(*board)]): # Vertical Combination
        return 'X'
    else: # Does not meet any of the conditions 
        return None
    
print(tic_tac_toe(board1))  # X
print(tic_tac_toe(board2))  # X
print(tic_tac_toe(board3))  # O
print(tic_tac_toe(board4))  # X
print(tic_tac_toe(board5))  # O
print(tic_tac_toe(board6))  # None
print(tic_tac_toe(board7))  # O
print(tic_tac_toe(board8))  # O
print(tic_tac_toe(board9))  # X
```

```python
# Problem 3: Routing (9 points)
# During the pandemic, a shuttle van service is tasked to travel along a predefined circular route as follows:

# UP Diliman -> Ateneo de Manila (Estimated Time: 10 mins)
# Ateneo de Manila -> De La Salle Taft (Estimated Time: 35 mins)
# De La Salle Taft -> UP Diliman (Estimated Time: 55 mins)
# The route is one-way only. So, the van cannot go back directly to UP Diliman from Ateneo de Manila; rather, it will have to drop by De La Salle Taft first before going to UP Diliman.


# Write a function called eta that takes three positional arguments: (dict) legs, (str) source, and (str) destination. The function should return the estimated number of minutes it will take to reach the destination from the source based on the data stored in the legs dictionary.

# Specifications:

# 1. The legs dictionary adheres to the schema followed in the demo cell.
# 2. There may be more legs than shown in the demo cell.
# 3. Any destination can be reached from any source.


# NON-INTERACTIVE CODE CELL. YOU MAY RUN THIS CELL, BUT DO NOT EDIT IT.
# FOR DEMONSTRATION PURPOSES ONLY. DO NOT EDIT.

legs = {
         ("upd","admu"):{
             "travel_time_mins":10
         },
         ("admu","slex"):{
             "travel_time_mins":125
         },
         ("slex","naia"):{
             "travel_time_mins":45
         },
         ("naia","dlsu"):{
             "travel_time_mins":35
         },
         ("dlsu","ust"):{
             "travel_time_mins":55
         },
         ("ust","nlex"):{
             "travel_time_mins":90
         },
         ("nlex","upd"):{
             "travel_time_mins":105
         },
    }
    

def eta(source, destination, legs):
    list_of_all_tuple_keys = list(legs.keys()) # Create a list consisting of tuple keys
    vehicle_at_destination = False # Vehicle has not yet arrived at the destination
    vehicle_route = [] # Create a new list vehicle_route
    
    if (str(source),str(destination)) in legs.keys(): # Inputs exactly match the data in the dictionary
        total_time_to_destination = legs[str(source),str(destination)]['travel_time_mins'] # Output eta based on the dictionary data
        return total_time_to_destination
    
    elif (str(source) == str(destination)): # Source and destination are equal
        return f'0'
    
    else:
        for location in (2 * list_of_all_tuple_keys):
            while not vehicle_at_destination: # While the vehicle is not yet at the destination
                if (location[0] == str(source)):
                    if (location[1] != str(destination)): # What if str(destination) is 'first than' str(source) on dictionary
                        vehicle_route += [location]
                elif (location[1] == str(destination)): # Combination of first and second
                    if (len(vehicle_route) >= 1):
                        vehicle_route += [location]
                        vehicle_at_destination = True # Vehicle has arrived at the destination
                elif (len(vehicle_route) > 0): # Going back to the top of the dictionary
                    vehicle_route += [location]
                break
        
    specific_route_to_be_taken = {} # Create a new set specific_route_to_be_taken
    for location, time_to_destination in legs.items():
        if location in vehicle_route:
            specific_route_to_be_taken[location] = time_to_destination # Add the tuple

    estimated_time_to_destination_route = [] # Create a new list estimated_time_to_destination_route
    for vehicle_route_value in specific_route_to_be_taken.values():
        estimated_time_to_destination_route += [vehicle_route_value] # Add the numerical value to the list

    total_time_to_destination = 0 # Set total_time_to_destination and increment the value
    for time_to_location in estimated_time_to_destination_route:
        total_time_to_destination += list(time_to_location.values())[0] # Index 0 is the first item in the list
    
    return total_time_to_destination
    
print(eta('dlsu','admu', legs)) # 260
print(eta('upd','dlsu', legs))  # 215
print(eta('admu','upd', legs))  # 455
print(eta('upd','ust', legs))   # 270
print(eta('upd','admu', legs))  #  10
print(eta('ust','upd', legs))   # 195
print(eta('ust','ust', legs))   #   0
print(eta('upd','nlex', legs))  # 360
print(eta('naia','slex', legs)) # 420
print(eta('admu','nlex', legs)) # 350
print(eta('dlsu','slex', legs)) # 385

# Reference: https://www.geeksforgeeks.org/shortest-path-problem-between-routing-terminals-implementation-in-python/
```
